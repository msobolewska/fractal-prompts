```python
# Parameters

# Length constraints from https://distantreading.github.io/Schema/eltec-1.html#index.xml-body.1_div.2_div.2 
short_min = 10000
short_max = 50000

medium_min = 50000
medium_max = 100000

long_min = 100000
long_max = "infinity"

# Prompts templates
prompt_basic = "Please generate a novel that has length in words bettwen {} and {}."

prompt_style = "Please generate a novel that has length in words bettwen {} and {}. Follow the style of {} which example works are {}."

prompt_parts = "Please generate the next chapter of a novel. The total length of the novel in words is bettwen {} and {}. Novel consists of following text so far: {}. Follow the style of {} which example works are {}."

# Directory with cloned repository https://github.com/COST-ELTeC/ELTeC-eng
eltec_dir = "."
```

```python
# Sample prompt
import xml.etree.ElementTree as ET

tree = ET.parse(eltec_dir + '/ELTeC-eng/level1/' + 'ENG18400_Trollope.xml')
namespace = {"ns": "http://www.tei-c.org/ns/1.0"}
root = tree.getroot()

body = root.find(".//ns:body", namespace)
body_raw_content = ET.tostring(body, encoding="unicode")

author = root.find(".//ns:author", namespace).text

sample_prompt_basic = prompt_basic.format(long_min, long_max)
sample_prompt_style = prompt_style.format(long_min, long_max, author, body_raw_content)
sample_prompt_parts = prompt_parts.format(long_min, long_max, '', author, body_raw_content)

print(sample_prompt_basic)
print(sample_prompt_style)
print(sample_prompt_parts)
```
